```python
import seaborn as sns
sns.set()
sns.set(style="darkgrid")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = pd.read_csv('https://sulagna206.github.io/Assignment/bigmart_data.csv')
```

```python
data1.head()
```

```python
data1.describe(include='all')
```

```python
data = data1.head(6)
data
```

```python
sns.countplot(x=data['Outlet_Location_Type'])
plt.show()
```

```python
sns.lineplot(x=data['Item_MRP'],y=data['Item_Outlet_Sales'],hue=data['Item_Fat_Content'],data=data)
plt.show()
```

```python
sns.barplot(x=data['Item_Type'],y=data['Item_MRP'],data=data)
plt.show()
```

```python
sns.distplot(data['Item_Outlet_Sales'])
plt.show()
```

```python
sns.boxplot(y=data['Item_MRP'],data=data)
plt.show()
```

```python
sns.boxplot(x=data['Outlet_Location_Type'],y=data['Item_Outlet_Sales'],data=data)
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.violinplot(y=data['Item_Outlet_Sales'],data=data,color='lightpink')
plt.show()
```

```python
sns.scatterplot(x=data1['Item_MRP'],y=data1['Item_Outlet_Sales'],hue=data1['Outlet_Location_Type'],data=data1)
plt.show()
```

```python
sns.scatterplot(x=data1['Item_MRP'][:100], y=data1['Item_Outlet_Sales'][:100], data=data1[:100], size=data1['Item_Fat_Content'][:100], hue=data1['Item_Fat_Content'][:100]);
```

```python
sns.catplot(x=data1['Item_Fat_Content'],y=data1['Item_MRP'],kind='strip',data=data1)
plt.show()
```
