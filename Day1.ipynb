```python
iris = pd.read_csv('https://sulagna206.github.io/Assignment/iris.csv')
iris_small=iris[:10]
iris_small
```

```python
plt.plot(iris_small['sepal_length'],iris_small['petal_length'])
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.title('Variation of petal length with sepal length')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(iris_small['sepal_length'],iris_small['petal_length'])
plt.plot(iris_small['sepal_length'],iris_small['sepal_width'])
plt.title('Variation of petal length and sepal width vs sepal length')
plt.legend(labels=['petal_length','sepal_width'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(iris_small['sepal_length'],iris_small['petal_length'],color='r')
plt.plot(iris_small['sepal_length'],iris_small['sepal_width'],'y--')
plt.title('Variation of petal length and sepal width vs sepal length')
plt.legend(labels=['petal_length','sepal_width'])
plt.show()
```

```python
group = iris.groupby(iris['species'],as_index=False).mean()
plt.plot(group['species'],group['petal_width'])
plt.plot(group['species'],group['petal_length'])
plt.legend(labels=['Petal width','Petal length'])
plt.xlabel('Species')
plt.ylabel('Value')
plt.title('Variation of petal length and width vs Species')
plt.show()
```

```python
sorted_iris = iris.sort_values('petal_length',ascending=False)
plt.bar(sorted_iris['species'],sorted_iris['petal_length'],color='red')
plt.title('Petal length for each species')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(group['species'],group['petal_length'])
plt.title('Variation of mean petal length')
plt.subplot(1,2,2)
plt.bar(group['species'],group['petal_length'],color=['red','blue','green'])
plt.title('Variation of mean petal length')
plt.show()
```
