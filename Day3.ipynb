```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv('https://sulagna206.github.io/Assignment/bigmart_data.csv')
```

```python
data.head()
```

```python
fig = px.line(data[:5],x='Item_Type',y='Item_MRP',color='Item_Fat_Content')
fig.show()
```

```python
fig = px.bar(data[:20],x='Item_Fat_Content',y='Item_MRP',color='Item_Type')
fig.show()
```

```python
fig = px.bar(data[:20],x='Outlet_Location_Type',y='Item_Outlet_Sales',color='Outlet_Type')
fig.show()
```

```python
px.bar(data[:20],y='Item_Outlet_Sales',x='Item_Type').show()
```

```python
px.histogram(data[:20],x='Item_MRP',color='Item_Type').show()
```

```python
px.histogram(data[:20],x='Item_Outlet_Sales',color='Item_Fat_Content')
```

```python
px.scatter(data,x='Item_MRP',y='Item_Outlet_Sales',color='Outlet_Location_Type').show()
```

```python
px.pie(data[:6],values='Item_MRP',names='Item_Type').show()
```

```python
px.box(data,x='Outlet_Location_Type',y='Item_Outlet_Sales').show()
```

```python
px.line_3d(data[:6],x='Item_MRP',y='Item_Weight',z='Item_Outlet_Sales').show()
```
